```python
import pandas as pd
import json
```

```python
df1 = pd.read_csv('population_total.csv', index_col=0)
df2 = pd.read_csv('life_expectancy_years.csv', index_col=0)
df3 = pd.read_csv('child_mortality_0_5_year_olds_dying_per_1000_born.csv', index_col=0)
df4 = pd.read_csv('total_gdp_us_inflation_adjusted.csv', index_col=0)
df6 = pd.read_csv('children_per_woman_total_fertility.csv', index_col=0)
```

```python
final_data = {}
years = df1.columns

for year in df1.columns:
    final_data[year] = {}
    for region in regions:
        final_data[year][region] = []
```

```python
for index, row in df1.iterrows():
    d1 = row
    d2 = df2.loc[index] if (index in df2.index) else pd.Series("undefined", index=years)
    d3 = df3.loc[index] if (index in df3.index) else pd.Series("undefined", index=years)
    d4 = df4.loc[index] if (index in df4.index) else pd.Series("undefined", index=years)
    d6 = df6.loc[index] if (index in df6.index) else pd.Series("undefined", index=years)
    for year in years:
        attributes = {}
        attributes['population'] = ("NaN" if pd.isna(d1[year]) else str(d1[year])) if (year in d1) else "undefined"
        attributes['lifeexpectancy'] =  ("NaN" if pd.isna(d2[year]) else str(d2[year])) if (year in d2) else "undefined"
        attributes['mortality'] = ("NaN" if pd.isna(d3[year]) else str(d3[year])) if (year in d3) else "undefined"
        attributes['gdp'] = ("NaN" if pd.isna(d4[year]) else str(d4[year])) if (year in d4) else "undefined"
        attributes['fertility'] = ("NaN" if pd.isna(d6[year]) else str(d6[year])) if (year in d6) else "undefined"
        country_year_data = {}
        country_year_data['country'] = index
        country_year_data['data'] = attributes
        #country_year_data[index] = attributes
        if index in df5.index:
            country_year_data['geo'] = df5.loc[index][0]
            country_year_data['region'] = df5.loc[index][1]
            final_data[year][df5.loc[index][1]].append(country_year_data)
            
    
stats = {}

stats['population'] = str(df1.max().max())
stats['lifeexpectancy'] = str(df2.max().max())
stats['mortality'] = str(df3.max().max())
stats['gdp'] = str(df4.max().max())
stats['fertility'] = str(df6.max().max())

final_data['stats'] = stats

print(final_data['stats'])
print("Done==========")    
#print(final_data)
with open('data.json', 'w') as f:
    json.dump(final_data, f, sort_keys=True, indent=4)
```
